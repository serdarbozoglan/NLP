```python
def read_file(filepath):
    with open(filepath) as f:
        str_text = f.read()
        
        return str_text 
```

```python
read_file('moby_dick_four_chapters.txt')
```

```python
import spacy
nlp =spacy.load('en', disable=['parser', 'tagger', 'ner']) # to faster tokenization we dsiable other features
```

```python
nlp.max_length = 1198623 
```

```python
def seperate_punc (doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n']
```

```python
d = read_file('moby_dick_four_chapters.txt')
```

```python
tokens = seperate_punc(d)
```

```python
tokens
```

```python
len(tokens)
```

```python
## we will feed 25 words and will predict 26th word (target word)
## 25 words --> network 26. word
## 25 word might be enough to predict the next word, less word may not be enough


train_len = 25 + 1
text_sequences = []

for i in range(train_len, len(tokens)):
    sequence = tokens[i-train_len:i]
    
    text_sequences.append(sequence)
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(text_sequences)
```

```python
sequences = tokenizer.texts_to_sequences(text_sequences)
```

```python
from keras.utils import to_categorical
```

```python
X = sequences[:, :-1] # Grab every column but last one
X
```

```python
y = sequences[:,-1] # grabs the very last column
y
```

```python
y = to_categorical(y, num_classes=vocabulary_size+1) # class number is vocab size+1 it needs hold 0 as well
```

```python
seq_len = X.shape[1] 
# seq_len grabs the number of 25 (which we set earlier)
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
```

```python
def create_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, seq_len, input_length=seq_len)) # input dimension, output dimension, input_length
    model.add(LSTM(seq_len*2, return_sequences=True)) # times of seq_len (2 times or 3 times etc, for this case 25X)
    model.add(LSTM(seq_len*2))
    model.add(Dense(50, activation='relu'))
    
    model.add(Dense(vocabulary_size, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.summary()
    
    return model
```

```python
model = create_model(vocabulary_size+1, seq_len)
```

```python
from pickle import dump, load
```

```python
model.fit(X, y, batch_size=128, epochs=2, verbose=1)
```

```python
model.save('my_mobydick_model.h5')
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    
    output_text = []
    input_text = seed_text
    
    for i in range(num_gen_words):
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        
        pad_encoded = pad_sequences([encoded_text], maxlen = seq_len, truncating='pre') # if tis is more than 25 words truncate the first words (pre words)
        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0] # brings the index of most probably word
        pred_word = tokenizer.index_word[pred_word_ind]
        
        input_text = ' '+pred_word
        output_text.append(pred_word)
    
    return ' '.join(output_text)
    
```
